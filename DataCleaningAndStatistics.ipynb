```python
df['idNumbers'] = df['Unique ID'].str.extract(r'ID_(\d+)').astype('int64')
df.set_index('idNumbers',inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
# labelencoder = LabelEncoder()
```

```python
cols = ['Unique ID','Suburb','Address','Type','Price','Method','SellerG','Date','Distance','Car','BuildingArea','YearBuilt','CouncilArea','Longtitude','Lattitude','Regionname']
def label_encode_columns(dff, columns):
    for col in columns:
        le = LabelEncoder().fit(dff[col])
        dff[col] = le.transform(dff[col])
    return dff

df2 = df.copy();
df2.dropna(inplace=True)
t_df = label_encode_columns(df2,cols)
t_df
```
