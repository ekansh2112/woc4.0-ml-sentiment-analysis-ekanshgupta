```python
import nltk
nltk.download('all')
```

```python

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re


def porterStemmer(text):
  ps = PorterStemmer()
  st = []
  for w in text:
    # print(w, " : ", ps.stem(w))
    st.append(ps.stem(w))
  
  return st

regex = re.compile('[^a-zA-Z]')
def clean_text(text):
    text = str(text).lower()
    text = regex.sub(' ', text)
    text = re.sub(' +', ' ', text)
    text = re.sub(r'^https?:\/\/.*[\r\n]*', '', text, flags=re.MULTILINE)
    return text

def tokenize(text):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]

    return filtered_sentence

df["Converted_Rating"] = df["Rating"].apply(lambda x: 0 if (x <= 2 or not x) else 1 if (x == 3) else 2)

df["Clean_Review"] = df["Review"].apply(lambda x: clean_text(x))

df["Tokenized_Review"] = df["Clean_Review"].apply(lambda x: tokenize(x))

df["Porter_Stemmer"] = df["Tokenized_Review"].apply(lambda x: porterStemmer(x))



# print(df.head(10))
```

```python
df["Review"] = df["Porter_Stemmer"]
df.head(3)
```
